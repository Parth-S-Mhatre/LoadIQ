```python
import pandas as pd

df = pd.read_csv("../DATA/time_series_30min_singleindex.csv")
```

```python
df["utc_timestamp"] = pd.to_datetime(df["utc_timestamp"])
df = df.sort_values("utc_timestamp")
df.set_index("utc_timestamp", inplace=True)
```

```python
df.info()
df.describe().T
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Value Pattern")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df.index, df["GB_GBN_load_actual_entsoe_transparency"])
plt.title("GB Load (Actual)")
plt.ylabel("MW")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df.index, df["GB_GBN_load_actual_entsoe_transparency"], label="Actual")
plt.plot(df.index, df["GB_GBN_load_forecast_entsoe_transparency"], label="Forecast")
plt.legend()
plt.title("Actual vs Forecast Load")
plt.show()
```

# Feature Engineering

```python
df["hour"] = df.index.hour
df["day_of_week"] = df.index.dayofweek
df["month"] = df.index.month
df["is_weekend"] = df["day_of_week"].isin([5,6]).astype(int)
```

```python
df["load_diff"] = (
    df["GB_GBN_load_actual_entsoe_transparency"] -
    df["GB_GBN_load_forecast_entsoe_transparency"]
)
```

```python
#Data Cleaning
```

```python
df = df.interpolate(method="time")
df = df.fillna(method="bfill").fillna(method="ffill")
```

Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler

features = [
    "GB_GBN_load_actual_entsoe_transparency",
    "GB_GBN_load_forecast_entsoe_transparency",
    "GB_GBN_solar_generation_actual",
    "GB_GBN_wind_generation_actual",
    "hour",
    "day_of_week",
    "load_diff"
]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])
```
